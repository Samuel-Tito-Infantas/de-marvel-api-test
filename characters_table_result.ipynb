```python
import os
import pandas as pd

from src.parameters import param_setted
```

```python
path = os.path.join(param_setted.get("output_target_folder"),"result","characters")

df_characters_comics = pd.read_parquet(os.path.join(path, "characters_comics_table.parquet"))
df_characters = pd.read_parquet(os.path.join(path, "characters_table.parquet"))
```

```python
df_characters_comics.head()
```

```python
df_characters.head()
```

```python
def create_result_file(df_characters_comics, df_characters):
    result = df_characters.merge(df_characters_comics,
                         on = "characters_id"
                         , how="left").rename(columns={
                            "comics_available":"quantity_of_comics_appear",
                            "characters_name":"Character_name"})
     
    result = result.sort_values(by = ["Character_name"] ,ascending=True)
    return result 
```

```python
result = create_result_file(df_characters_comics, df_characters)
result.head()
```

```python
path = os.path.join(param_setted.get("output_target_folder"), "result", "characters", "technical_result.csv")
columns = ["Character_name", "quantity_of_comics_appear"] 

result[columns].to_csv(path, sep=";",index=False)
```
